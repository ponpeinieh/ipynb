### Random walks example
- a path that is determined by a series of random decisions(for its direction)
- The Wiener process (or **Brownian Motion**) is an example of a stochastic process in continuous time. 
   - In 1828 the Scottish botanist **Robert Brown** observed that pollen grains suspended in water moved in an apparently random way, changing direction continuously, which was due to the pollen grains being bombarded by water molecules. 
   - The precise mathematical formulation of this phenomenon was later given by Norbert Wiener in 1923.

### RandomWalk Class
1. 3 attributes: 
   - number of points
   - a list of x-coordinates
   - a list of y-coordinates
2. 2 methods:
   - `__init__()` - ctor
   - `fill_walk()` - calculate the points in the walk

````python
from random import choice
class RandomWalk:
    ''' A class to generate random walks '''
    def __init__(self, num_points = 5000):
        ''' Initialize attributes '''
        self.num_points = num_points
        # all walks starts at origin 
        self.x_values=[0]
        self.y_values=[0]
    def fill_walk(self):
        ''' calculate all points in the walk'''
        while len(self.x_values)<self.num_points:
            x_direction = choice([1,-1])
            x_distance = choice(list(range(0,5)))
            x_step = x_direction*x_distance
            y_direction = choice([1,-1])
            y_distance = choice(list(range(0,5)))
            y_step = y_direction*y_distance
            # skip those not moving at all
            if x_step==0 and y_step==0:
                continue
            x = self.x_values[-1]+x_step
            y = self.y_values[-1]+y_step
            self.x_values.append(x)
            self.y_values.append(y)
# create RandomWalk instance
rw = RandomWalk()
rw.fill_walk()
# Plot the points in the walk.
plt.style.use('classic')
fig, ax = plt.subplots()
ax.scatter(rw.x_values, rw.y_values, s=5)
plt.show()
````

````python
# Keep making new walks, as long as the program is active.
while True:
    # Make a random walk.
    rw = RandomWalk()
    rw.fill_walk()
    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s=5)
    plt.show()
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
````

### Style the walk
1. emphasize the important characteristics of each walk - eg. where the walk started, where it ended and the path taken
2. deemphasize distracting elements - eg. tick marks and labels
3. coloring the points :
   - use a colormap to show the order of the points in the walk
   - remove the black outline from each dot
   - To color the points according to their position in the walk, we pass the c argument a list containing the position of each point. Because the points are plotted in order, the list just contains the numbers from 0 to 4999
   

````python
rw = RandomWalk()
rw.fill_walk()
# Plot the points in the walk.
plt.style.use('classic')
fig, ax = plt.subplots()
point_numbers = range(rw.num_points) # 0 to 4999
ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap="Blues", s=5, edgecolors = 'none')
plt.show()
````

### Analysis of rolling die
- Die class - simulate the roll of one die

````python
from random import randint
class Die:
    '''A class representing a single die'''
    def __init__(self, num_sides=6):
        '''Assume a 6-side die'''
        self.num_sides = num_sides
    def roll(self):
        '''return a random integer between 1 and number of sides'''
        return randint(1,self.num_sides) # note that the ending value is num_sides
````

````python
# roll the die 100 times and see the results
die = Die()
results=[]
for roll in range(100):
    results.append(die.roll())
print(results)
````

````python
#Analyze results
die = Die()
results=[]
for roll in range(1000):
    results.append(die.roll())
# analyze results
frequencies = [0]*die.num_sides
for r in results:
    frequencies[r-1]+=1
print(frequencies)
````

````python
#Analyze results
die = Die()
results=[]
for roll in range(1000):
    results.append(die.roll())
# analyze results
frequencies = []
for i in range(1,die.num_sides+1):
    frequencies.append(results.count(i))
print(frequencies)
````

````python
# Draw a histogram
from plotly.graph_objs import Bar, Layout
from plotly import offline

# Visualize the results.
x_values = list(range(1, die.num_sides+1)) # transform range to list
# Bar() create a bar instance that represents a data set formatted as a bar chart
# a data set may contain many elements, so Bar() needs to be wrapped in []
data = [Bar(x=x_values, y=frequencies)] 

x_axis_config = {'title': 'Result'} # a dict
y_axis_config = {'title': 'Frequency of Result'} # a dict
#Layout() return an instance that specifies the layout and config of the graph as a whole
my_layout = Layout(title='Results of rolling one D6 1000 times', xaxis=x_axis_config, yaxis=y_axis_config)
offline.plot({'data': data, 'layout': my_layout}, filename='d6.html')
````

````python
# roll two dice
die_1 = Die()
die_2 = Die()
results=[]
for roll in range(1000):
    results.append(die_1.roll()+die_2.roll())
# analyze results
frequencies = []
for i in range(2,die_1.num_sides+die_2.num_sides+1): # from 2 to 12
    frequencies.append(results.count(i))
#print(frequencies)
# Visualize the results.
x_values = list(range(2, die_1.num_sides+die_2.num_sides+1)) 
data = [Bar(x=x_values, y=frequencies)] 

x_axis_config = {'title': 'Result'}
y_axis_config = {'title': 'Frequency of Result'} 
my_layout = Layout(title='Results of rolling two D6 1000 times', xaxis=x_axis_config, yaxis=y_axis_config)
offline.plot({'data': data, 'layout': my_layout}, filename='2d6.html')
````
