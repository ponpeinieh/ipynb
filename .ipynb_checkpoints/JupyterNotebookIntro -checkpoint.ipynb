# Jupyter notebook

## 什麼是 Jupyter notebook?

- 一種開發及呈現Data Science Projects的互動工具

## 什麼是 notebook?

- Notebook(筆記本)整合了執行程式碼與呈現輸出的一種文件格式, 輸出形式包含了資料視覺化呈現, 說明文件 , 數學公式, 及其他媒體呈現等形式
- Notebook中可以執行多種不同的程式語言的程式, 但是大多數是使用Python語言

## cell

1. 目前正在聚焦的cell, 其外框顏色有:

   - 藍色: 代表命令(command)模式(按下esc進入)
   - 綠色: 代筆編輯(edit)模式(按下enter進入)
   
2. 命令模式可以:

   - 按下Ctrl + Enter 可以執行cell
   - 按下A: 在上方插入新的cell
   - 按下B: 在下方插入新的cell
   - Up, Down 按鍵可以上, 下移動到鄰近的cells
   - 按下M : 轉換成Markdown text cell
   - 按下Y : 轉換成code cell
   - 按下DD: 刪除目前的cell
   - 按下Z: 復原刪除
   - 按下Shift+Up或Down鍵: 可以多選cells, 然後做合併成一個cell(按Shift+M)
      - 也可以Shift+mouse Click cell左邊的空白地區,做多選cells
   

````python
#先在編輯模式, 利用 Ctrl+Shift+Minus做split, 在試着merge
message="Hello World"
print(message)

message="Hello World2"
print(message)

message="Hello World3"
print(message)

message="Hello World4"
print(message)
````

## kernel: notebook背後的執行環境
1. kernel執行完某個cell後, 將結果回傳給cell做呈現
2. kernel會保持狀態, 例如:
   - 在某個cell中宣告的變數, 可以在編號之後的cells裏使用
   - 在某個cell中import的libraries或定義的函數, 可以在編號之後的cells裏使用
   - 備注: 
      - **與cell放置的上下順序無關, 只要cell已經被執行過**
      - 注意cells可能會被修改然後重新執行, 所以放置的上下順序不是真正的執行順序, In [ ]中的編號才是執行的順序

````python
# 使用之前cell宣告的message變數
print(message)
````

3. 可以Restart kernel, 有下面三種模式:
   - **Restart** : 重新啓動kernel, 並且清除所有變數
   - **Restart & Clear Output** : 重新啓動kernel, 清除所有變數, 並且清除所有cell下方的輸出
   - **Restart & Run All**: 重新啓動kernel, 清除所有變數, 並且清除所有cell下方的輸出, 然後由上而下依序重新執行所有的cells
