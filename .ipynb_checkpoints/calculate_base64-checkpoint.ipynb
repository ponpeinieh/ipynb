```python
def fromBase64String(stringObject):
    result=b''
    length = len(stringObject)
    start=0
    while start<length:
        chunk = stringObject[start:start+4]
        result += chunkToBytes(chunk)
        start+=4
    print(result)
def toBase64String(bytesobject):
    result=''
    length = len(bytesobject) 
    start=0
    while start<length:
        chunk = bytesobject[start:start+3]
        result += chunkToString(chunk)
        start+=3
    print(result)
def mapBase64CharToInt(c):
    if c=='/':
        return 63
    elif c=='+':
        return 62
    elif c>='0' and c<='9':
        return 52+ord(c)-ord('0')
    elif c>='a' and c<='z':
        return 26+ord(c)-ord('a')
    elif c>='A' and c<='Z':
        return ord(c)-ord('A')
def mapIntToBase64Char(i):
    if i<26:
        return chr(ord('A')+i)
    elif i<52:
        return chr(ord('a')+i-26)
    elif i<62:
        return chr(ord('0')+i-52)
    elif i==62:
        return '+'
    elif i==63:
        return '/'
def chunkToBytes(stringchunk): # 4 characters
    result=b''
    if "==" in stringchunk:
        data = [mapBase64CharToInt(c) for c in stringchunk[:2]]
        # get first byte
        temp1 = (data[0]<<2) + (data[1]>>4)
        result+=temp1.to_bytes(1, 'big')
    elif "=" in stringchunk:
        data = [mapBase64CharToInt(c) for c in stringchunk[:3]]
        # get first byte
        temp1 = (data[0]<<2) + (data[1]>>4)
        result+=temp1.to_bytes(1, 'big')
        # get second byte
        temp1 = ((data[1]&15)<<4) + (data[2]>>2)
        result+=temp1.to_bytes(1, 'big')
    else:
        data = [mapBase64CharToInt(c) for c in stringchunk]
        # get first byte
        temp1 = (data[0]<<2) + (data[1]>>4)
        result+=temp1.to_bytes(1, 'big')
        # get second byte
        temp1 = ((data[1]&15)<<4) + (data[2]>>2)
        result+=temp1.to_bytes(1, 'big')
        # get third byte
        temp1 = ((data[2]&3)<<6) + data[3]
        result+=temp1.to_bytes(1, 'big')
    return result
def chunkToString(bytechunk): #maximum 3 bytes
    result = ''
    if len(bytechunk) == 3:   #check if input is of 3-byte long
        #so total 24 bits (0-23)
        #get bit 0-5
        result+=mapIntToBase64Char(bytechunk[0]>>2)

        #get bit 6-11
        temp1 = (bytechunk[0] & 3) << 4
        temp2 = bytechunk[1] >>4
        result+=mapIntToBase64Char(temp1+temp2) 

        #get bit 12-17
        temp1 = (bytechunk[1] & 15) << 2
        temp2 = bytechunk[2] >>6
        result+=mapIntToBase64Char(temp1+temp2) 

        #get bit 18-23
        temp1 = bytechunk[2] & 63
        result+=mapIntToBase64Char(temp1) 
    elif len(bytechunk)==2:
        #get bit 0-5
        result+=mapIntToBase64Char(bytechunk[0]>>2)

        #get bit 6-11
        temp1 = (bytechunk[0] & 3) << 4
        temp2 = bytechunk[1] >>4
        result+=mapIntToBase64Char(temp1+temp2) 
        
        #bit 12-17
        temp1 = (bytechunk[1] & 15) << 2
        result+=mapIntToBase64Char(temp1) 
        
        #bit 18-23 => '='
        result+='='
    elif len(bytechunk)==1:
        #get bit 0-5
        result+=mapIntToBase64Char(bytechunk[0]>>2)

        #get bit 6-11
        temp1 = (bytechunk[0] & 3) << 4
        result+=mapIntToBase64Char(temp1) 
        
        #bit 12-17 and 18-23 => '=='  
        result+='=='
    return result
toBase64String(b'Many hands make light wor')
fromBase64String('TWFueSBoYW5kcyBtYWtlIGxpZ2h0IHdvcg==')
```
