```python
# Depth-First Search(DFS) traversal : inorder
# define a Node (節點) class 
class Node:
    def __init__(self,value):
        self.value = value # 儲存的數值
        self.left = None
        self.right = None
def traverseDFSInOrder(root):
    if root:
        traverseDFSInOrder(root.left)
        visitNode(root)
        traverseDFSInOrder(root.right)
def traverseDFSPreOrder(root):
    if root:
        visitNode(root)
        traverseDFSPreOrder(root.left)
        traverseDFSPreOrder(root.right)
def traverseDFSPostOrder(root):
    if root:
        traverseDFSPostOrder(root.left)
        traverseDFSPostOrder(root.right)
        visitNode(root)

inOrderData = [10,20,30,40,50,60,70,80]
preOrderData = [50,30,20,10,40,70,60,80]
preOrderIndex=0
root_node = createTree(0,len(inOrderData)-1)

traverseDFSInOrder(root_node)
print()
traverseDFSPreOrder(root_node)
print()
traverseDFSPostOrder(root_node)
```

```python
# DFS preorder traversal : an iterative approach using a stack
def dfs(root):
    if root is None:
        return
    stack = [root]
    while stack:
        parent = stack.pop()
        if parent.right:
            stack.append(parent.right)
        if parent.left:
            stack.append(parent.left)
        visitNode(parent)

inOrderData = [10,20,30,40,50,60,70,80]
preOrderData = [50,30,20,10,40,70,60,80]
preOrderIndex=0
root_node = createTree(0,len(inOrderData)-1)

dfs(root_node)
```
