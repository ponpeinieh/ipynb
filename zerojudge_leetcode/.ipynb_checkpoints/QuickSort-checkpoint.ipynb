```python
import random
import tracemalloc
 
def quick_sort_simple(A):
    if not A or len(A)<=1:
        return A
    pivot = A[-1]
    Al=[]
    Ar=[]
    for i in range(len(A)-1):
        if A[i]<=pivot:
            Al.append(A[i])
        else:
            Ar.append(A[i])
    return quick_sort_simple(Al)+[pivot]+quick_sort_simple(Ar)

tracemalloc.start()
length=20
A = random.sample(range(length),k=length)
print("Before sort", A)
B = quick_sort_simple(A)
print("After sort", B) 

# displaying the memory
print(tracemalloc.get_traced_memory()) 

tracemalloc.reset_peak()
length=40
A = random.sample(range(length),k=length)
print("Before sort", A)
B = quick_sort_simple(A)
print("After sort", B) 

# displaying the memory
print(tracemalloc.get_traced_memory())
# stopping the library
tracemalloc.stop()                
```

```python
def find_pivot_index(A,start,end):
    pivot=A[end]
    p=-1
    for i in range(start,end):
        if A[i] <= pivot and p!=-1:
            A[p],A[i]=A[i],A[p]
            p+=1
        elif A[i]>pivot and p==-1:
            p=i
    if p!=-1:
        A[p],A[end]=pivot,A[p]
        return p
    else:
        return end
    
def find_pivot_index_2(A,start,end):
    pivot=A[end]
    p=start
    for i in range(start,end):
        if A[i] <= pivot:
            A[p],A[i]=A[i],A[p]
            p+=1
    A[p],A[end]=pivot,A[p]
    return p

import random
def quick_sort(A,start,end):
    if start<end:
        pivot = find_pivot_index_2(A,start,end)
        quick_sort(A,start,pivot-1)
        quick_sort(A,pivot+1,end)
        
length=20
A = random.sample(range(length),k=length)
print("Before sort", A)
quick_sort(A,0,length-1)
print("After sort", A) 
        
```

```python
def find_pivot_index_unstable(A,start,end):
    pivot=A[end]
    p=start
    for i in range(start,end):
        if A[i] <= pivot:
            A[p],A[i]=A[i],A[p]
            p+=1
        print(i, A)
    A[p],A[end]=pivot,A[p]
    return p

def find_pivot_index_stable(A,start,end):
    pivot=A[end]
    p=start
    for i in range(start,end):
        if A[i] <= pivot:
#             removeAndInsert(A, p, i)
            rightShiftAndReplace(A, p, i)
            p+=1
        print(i, A)
#     removeAndInsert(A,p,end)
    rightShiftAndReplace(A, p, end)
    return p

def removeAndInsert(A, p, i):
    #Method 1: pop element at position i and insert into position p
    e = A.pop(i)
    A.insert(p,e)
    
def rightShiftAndReplace(A, p, i):
    #Method 2: 'right' shift elements between p and i-1 by 1 and 
    # replace element at position p with position i
    i_element = A[i]
    for n in range(i, p, -1):
        A[n] = A[n-1]
    A[p]=i_element


    
import random 
length=10
A = random.sample(range(length),k=length)
CA = A[:]
find_pivot_index_unstable(A,0,length-1)
print(A)
print()
find_pivot_index_stable(CA,0,length-1)
print(CA)
```
