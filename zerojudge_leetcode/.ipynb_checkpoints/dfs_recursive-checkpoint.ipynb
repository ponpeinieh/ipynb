```python
# define a Node (節點) class 
class Node:
    def __init__(self,value):
        self.value = value # 儲存的數值
        self.left = None
        self.right = None

def visitNode(node):
    print(node.value, end=",")

def createTree(start_i, end_i): 
    # start_i, end_i: start and end indices for inorder data
    global preOrderIndex # next root index in preorder data 
    if start_i > end_i:
        return None
    node = Node(preOrderData[preOrderIndex]) #create root node
    preOrderIndex+=1
    # get the root node index position in inorder data
    inOrderIndex = inOrderData.index(node.value, start_i, end_i+1)
    # recursively create the left and right subtrees
    node.left = createTree(start_i, inOrderIndex-1)
    node.right = createTree(inOrderIndex+1, end_i)
    return node

def traverseDFSPreOrder(root):
    if root:
        visitNode(root)
        traverseDFSPreOrder(root.left)
        traverseDFSPreOrder(root.right)
```

```python
inOrderData = [10,20,30,40,50,60,70]
preOrderData = [40,20,10,30,60,50,70]
preOrderIndex=0
root_node = createTree(0, len(inOrderData)-1)
traverseDFSPreOrder(root_node)
print()
```

```python
def dfs(graph, root_node):
    traversed = [root_node]
    stack = [root_node]
    while stack:
        vertex = stack[-1]
        if vertex not in traversed:
            traversed.append(vertex)
        pop = True
        for adjacent in graph[vertex]:
            if adjacent not in traversed:
                stack.append(adjacent)
                pop = False
                break
        if pop:
            stack.pop()
    return traversed


graph = {"1": ["5", "9"], "5": ["2", "4"], "9": ["8"], "2": ["4"], "4": ["2"], "8": []}
print(dfs(graph, "1"))
```
