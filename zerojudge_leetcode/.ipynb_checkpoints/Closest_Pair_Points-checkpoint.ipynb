```python
import math
import sys
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __str__(self):
        return f'x:{self.x}, y:{self.y}'
    def distanceTo(self, p):
        return math.sqrt((self.x-p.x)**2+(self.y-p.y)**2)
    
def findClosestPairBruteForce(Px, start, end):
    closest_pair_dist=[]
    closest_dist = sys.float_info.max
    for i in range(start, end):
        for j in range(i+1, end+1):
            d = Px[i].distanceTo(Px[j])
            if d<closest_dist:
                closest_dist=d
                closest_pair_dist=[Px[i],Px[j], closest_dist]
    return closest_pair_dist

def findClosestPairCross(Px, Py, middle, cpd):
    strip=[]
    m_x = Px[middle].x
    closest_dist = cpd[2]
    for p in Py:
        if p.x>= m_x-closest_dist and p.x<=m_x+closest_dist:
            strip.append(p)
    for i in range(len(strip)):
        for j in range(i+1, len(strip)):
            if strip[j].y - strip[i].y > closest_dist:
                break
            dist = strip[j].distanceTo(strip[i])
            if dist<closest_dist:
                closest_dist = dist
                cpd = [strip[i], strip[j], closest_dist]
                
    return cpd

def findClosestPair(Px, Py, start, end):
    if end-start<3:
        return findClosestPairBruteForce(Px, start, end)
    else:
        middle = start+(end-start)//2
        cpd_left = findClosestPair(Px, Py, start, middle)
        cpd_right = findClosestPair(Px, Py, middle+1, end)
        if cpd_left[2]<cpd_right[2]:
            cpd = cpd_left
        else:
            cpd = cpd_right
        cpd_cross = findClosestPairCross(Px, Py, middle, cpd)
        if cpd[2]<cpd_cross[2]:
            return cpd
        else:
            return cpd_cross
    
    
import random
def main():
    P = []
    sample_x = random.sample(range(500),k=10)
    sample_y = random.sample(range(500),k=10)
    for i in range(10):
        p = Point(sample_x[i], sample_y[i])
        P.append(p)
        print(p)
    #sort on x and y
    Px = sorted(P,key=lambda p:p.x)
    Py = sorted(P,key=lambda p:p.y)
    #bruteforce
    cpd = findClosestPairBruteForce(Px, 0, len(P)-1)
    print(f'The closest pair of points are: {cpd[0]}, {cpd[1]} with distance {cpd[2]}')
    cpd = findClosestPair(Px,Py,0,len(P)-1)
    print(f'The closest pair of points are: {cpd[0]}, {cpd[1]} with distance {cpd[2]}')
    
main()
```
