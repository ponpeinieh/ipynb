```python
# zerojudge m372. 3. 搬家 2023年10月APCS
def util():
    global current_c,max_c
    for ri in range(len(data)):
        for ci in range(len(data[ri])):
            if data[ri][ci]!='0' and data[ri][ci]!='C':
                current_c=0
                # dfs(ri,ci,None) # this times out !!!
                dfs_iterative(ri,ci,None) # this passed!!!
                if current_c>max_c:
                    max_c = current_c

def dfs_iterative(sri,sci,sid):
    global current_c,max_c
    stack = [(sri,sci,sid)]
    while stack:
        el = stack.pop()
        ri = el[0]
        ci = el[1]
        sid = el[2] # source input direction, R,L,U, or D
        va = data[ri][ci]
        if va=='0' or va=='C':
            continue
        current_c+=1
        data[ri][ci]='C' # mark as checked 'C'
        if sid!='R' and va in 'FHXL' and ci<m-1: # right
            nv = data[ri][ci+1]
            if nv in 'H7XJ': # has left input
                stack.append((ri,ci+1,'L'))
        if sid!='L' and va in 'H7XJ' and ci>0: # left
            nv = data[ri][ci-1]
            if nv in 'FHXL': # has right input
                stack.append((ri,ci-1,'R'))
        if sid!='U' and va in 'IXLJ' and ri>0: # up
            nv = data[ri-1][ci]
            if nv in 'F7IX': # has down input
                stack.append((ri-1,ci,'D'))
        if sid!='D' and va in 'F7IX' and ri<n-1: # down
            nv = data[ri+1][ci]
            if nv in 'IXLJ': # has up input
                stack.append((ri+1,ci,'U'))        
def dfs(ri,ci, sid):
    global current_c,max_c
    va = data[ri][ci]
    if va=='0' or va=='C': # 'C' means checked
        return
    current_c+=1
    data[ri][ci]='C'
    if va == 'F': # right and down
        if sid!='R' and ci<m-1 : # right
            nv = data[ri][ci+1]
            if nv in 'H7XJ':
                dfs(ri,ci+1,'L')
        if sid!='D' and ri<n-1: # down
            nv = data[ri+1][ci]
            if nv in 'IXLJ':
                dfs(ri+1,ci,'U')
        return
    elif va=='H': # left and right
        if sid!='L' and ci>0: # left
            nv = data[ri][ci-1]
            if nv in 'FHXL':
                dfs(ri,ci-1,'R')
        if sid!='R' and ci<m-1: # right
            nv = data[ri][ci+1]
            if nv in 'H7XJ':
                dfs(ri,ci+1,'L')
        return
    elif va == '7': # left and down
        if sid!='L' and ci>0: # left
            nv = data[ri][ci-1]
            if nv in 'FHXL':
                dfs(ri,ci-1,'R')
        if sid!='D' and ri<n-1: # down
            nv = data[ri+1][ci]
            if nv in 'IXLJ':
                dfs(ri+1,ci,'U')
        return  
    elif va == 'I': # up and down
        if sid!='U' and ri>0: # up
            nv = data[ri-1][ci]
            if nv in 'F7IX':
                dfs(ri-1,ci,'D')
        if sid!='D' and ri<n-1: # down
            nv = data[ri+1][ci]
            if nv in 'IXLJ':
                dfs(ri+1,ci,'U')
        return  
    elif va == 'X': # up, down, left and right
        if sid!='U' and ri>0: # up
            nv = data[ri-1][ci]
            if nv in 'F7IX':
                dfs(ri-1,ci,'D')
        if sid!='D' and ri<n-1: # down
            nv = data[ri+1][ci]
            if nv in 'IXLJ':
                dfs(ri+1,ci,'U')
        if sid!='L' and ci>0: # left
            nv = data[ri][ci-1]
            if nv in 'FHXL':
                dfs(ri,ci-1,'R')
        if sid!='R' and ci<m-1: # right
            nv = data[ri][ci+1]
            if nv in 'H7XJ':
                dfs(ri,ci+1,'L')
        return
    elif va == 'L': # up and right
        if sid!='U' and ri>0: # up
            nv = data[ri-1][ci]
            if nv in 'F7IX':
                dfs(ri-1,ci,'D')
        if sid!='R' and ci<m-1: # right
            nv = data[ri][ci+1]
            if nv in 'H7XJ':
                dfs(ri,ci+1,'L')
        return
    elif va == 'J': # up, left 
        if sid!='U' and ri>0: # up
            nv = data[ri-1][ci]
            if nv in 'F7IX':
                dfs(ri-1,ci,'D')
        if sid!='L' and ci>0: # left
            nv = data[ri][ci-1]
            if nv in 'FHXL':
                dfs(ri,ci-1,'R')
        return
    return

            
n,m = map(int,input().split())
mps = [input() for _ in range(n)]
# n,m=3,4
# mps=['FHH7','IIII','LHHJ']
# n,m=4,7
# mps=['0F70000','FXJ0000','II700X7','LJ0HHLJ']
data=[list(s) for s in mps]
#print(data)
current_c = 0
max_c = 0
util()
print(max_c)
```
