```python
def fill_zero(a,b):
    if len(a)>len(b):
        b='0'*(len(a)-len(b))+b
    elif len(a)<len(b):
        a='0'*(len(b)-len(a))+a
    return a,b
def resultToStr(result): # result is a list of single-digit characters, may have prefix 0 (remove them)
    for i in range(len(result)):
        if result[i]=='0':
            continue
        else:
            return "".join(result[i:])
    return '0'

def sum(a,b):
    if a=='0':
        return b
    elif b=='0':
        return a
    #a,b are integer strings, eg '123456', '897'
    #fill prefix-'0' , so a and b are of equal length
    a,b = fill_zero(a,b)
    carry=0
    result=['0']*len(a) 
    for i in range(len(a)-1, -1, -1):      
        s = int(a[i])+int(b[i])+carry
        result[i] = str(s%10)
        carry = s//10
    result.insert(0, str(carry))
    return resultToStr(result)

def subtract(a,b): #a,b are integer strings, eg '123456', '897', assume a may be less than b
    if a=='0':
        return '-'+b
    elif b=='0':
        return a
    a,b = fill_zero(a,b)
    negative = False
    if a<b: #string comparison
        negative = True
        #swap
        a,b = b,a
    borrow=0
    result=['0']*len(a) 
    for i in range(len(a)-1, -1, -1):      
        s = int(a[i])-int(b[i])+borrow
        if s<0:
            s+=10
            borrow = -1
        else:
            borrow = 0
        result[i] = str(s)
    if negative:
        return '-'+resultToStr(result)
    else:
        return resultToStr(result)
def multiply_simple(a,b): # assume a is an integer string, b is single-digit character
    if b=='0':
        return '0'
    carry=0
    result=['0']*len(a) 
    for i in range(len(a)-1, -1, -1):      
        s = int(a[i])*int(b)+carry
        result[i] = str(s%10)
        carry = s//10
    result.insert(0, str(carry))
    return resultToStr(result)
    
def multiply(a,b): #a,b are integer strings, eg '123456', '897' 
    if a=='0' or b=='0':
        return '0'
    if len(a)<len(b):
        a,b=b,a #ensure len(a) is larger than len(b)
    carry=0
    result='0'
    for i in range(len(b)-1, -1, -1):  #897
        r = multiply_simple(a, b[i])
        result = sum(result,r)
        a=a+'0'
    return result
    
a, op, b = [i for i in input().split()]
if op=='+':
    print(sum(a,b))
elif op=='-':
    print(subtract(a,b))
elif op=='*':
    print(multiply(a,b))
```

```python
def resultToStr(result): # result is a list of single-digits 
    for i in range(len(result)):
        if result[i]==0:
            continue
        else:
            return "".join([str(c) for c in result[i:]])
    return '0'
def is_zero(n): # n is a list of single digits
    if len(n)==0 or (len(n)==1 and n[0]==0):
        return True
    return False
def sum_str(a,b): # a,b are strings
    a = [int(i) for i in a]
    b = [int(i) for i in b]
    return resultToStr(sum(a,b))
def sum(a,b): # a,b are list of single digits
    if is_zero(a):
        return b
    elif is_zero(b):
        return a
    carry=0
    result=[0]*max(len(a),len(b))
    for i in range(max(len(a),len(b))):   
        a_i = len(a)-i-1
        b_i = len(b)-i-1
        if a_i<0:
            s = b[b_i]+carry
        elif b_i<0:
            s = a[a_i]+carry
        else:
            s = a[a_i]+b[b_i]+carry
        result[len(result)-i-1] = s%10
        carry = s//10
    if carry:
        result.insert(0, carry)
    return result

def compare(a,b):# a,b are strings
    if len(a)>len(b):
        return True
    elif len(a)<len(b):
        return False
    else:
        return a>=b
    
def subtract_str(a,b):  # a,b are strings
    negative=False
    if not compare(a,b):
        negative=True
        a,b = b,a #swap  
    a = [int(i) for i in a]
    b = [int(i) for i in b]
    if is_zero(a):
        return '0'
    elif is_zero(b):
        if negative:
            return '-'+resultToStr(a)
        else:
            return resultToStr(a)
    borrow=0
    result=[0]*max(len(a),len(b)) 
    for i in range(max(len(a),len(b))):     
        a_i = len(a)-i-1
        b_i = len(b)-i-1
        if a_i<0:
            s = b[b_i]+borrow
        elif b_i<0:
            s = a[a_i]+borrow
        else:
            s = a[a_i]-b[b_i]+borrow
        if s<0:
            s+=10
            borrow = -1
        else:
            borrow = 0
        result[len(result)-i-1] = s
    if negative:
        return '-'+resultToStr(result)
    else:
        return resultToStr(result)
def multiply_simple(a,b):  # a is a list of single digits,  b is a single-digit integer
    if b==0:
        return [0]
    carry=0
    result=[0]*len(a) 
    for i in range(len(a)):     
        a_i = len(a)-i-1
        s = a[a_i]*b+carry
        result[len(result)-i-1] = s%10
        carry = s//10
    if carry:
        result.insert(0, carry)
    return result
    
def multiply_str(a,b):  # a,b are strings
    if len(a)<len(b):
        a,b=b,a #ensure len(a) is larger than len(b)
    a = [int(i) for i in a]
    b = [int(i) for i in b]
    if is_zero(a) or is_zero(b):
        return '0'
    carry=0
    result=[0]
    for i in range(len(b)):  
        b_i = len(b)-i-1
        r = multiply_simple(a, b[b_i])
        if i>0:
            r = r + [0]*i
        result = sum(result,r)
    return resultToStr(result)
    
a, op, b = [i for i in input().split()]
if op=='+':
    print(sum_str(a,b))
elif op=='-':
    print(subtract_str(a,b))
elif op=='*':
    print(multiply_str(a,b))
```
