```python
import random
 
def quick_sort_simple(A):
    if not A or len(A)<=1:
        return A
    pivot = A[-1]
    Al=[]
    Ar=[]
    for i in range(len(A)-1):
        if A[i]<=pivot:
            Al.append(A[i])
        else:
            Ar.append(A[i])
    return quick_sort_simple(Al)+[pivot]+quick_sort_simple(Ar)


length=100
A = random.sample(range(length),k=length)
print("Before sort", A) 
B = quick_sort_simple(A)
print("after sort", B) 
```

```python
def find_pivot_index(A,start,end):
    pivot=A[end]
    p=-1
    for i in range(start,end):
        if A[i] <= pivot and p!=-1:
            A[p],A[i]=A[i],A[p]
            p+=1
        elif A[i]>pivot and p==-1:
            p=i
    if p!=-1:
        A[p],A[end]=pivot,A[p]
        return p
    else:
        return end
    
def find_pivot_index_2(A,start,end):
    pivot=A[end]
    p=start
    for i in range(start,end):
        if A[i] <= pivot:
            A[p],A[i]=A[i],A[p]
            p+=1
    A[p],A[end]=pivot,A[p]
    return p

import random
def quick_sort(A,start,end):
    if start<end:
        pivot = find_pivot_index_2(A,start,end)
        quick_sort(A,start,pivot-1)
        quick_sort(A,pivot+1,end)
   

length=100
A = random.sample(range(length),k=length) 
print("Before sort", A) 
quick_sort(A, 0, len(A)-1)
print("After sort", A) 
```

```python
def partition(A,start,end):
    mid = (end+start)//2
    pivot = A[mid]
    while not start>end:
        #'start' points to the next element that is not less than the pivotal
        while A[start] < pivot:
            start+=1
        #'end' points to the next element that is not larger than the pivotal
        while A[end] > pivot:
            end-=1
        #Swap elements and advance the pointers 'start' and 'end' by one 
        if not start>end: 
            A[start], A[end] = A[end], A[start] #swap
            start+=1
            end-=1 
    return start

import random
def quick_sort(A,start,end):
    if start<end:
        pivot = partition(A,start,end)
        quick_sort(A,start,pivot-1)
        quick_sort(A,pivot,end)
        
length=12
A = random.sample(range(length),k=length)
print("Before sort", A)
quick_sort(A,0,length-1)
print("After sort", A) 




```
