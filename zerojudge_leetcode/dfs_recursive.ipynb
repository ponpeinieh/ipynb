```python
# define a Node (節點) class 
class Node:
    def __init__(self,value):
        self.value = value # 儲存的數值
        self.left = None
        self.right = None

def visitNode(node):
    print(node.value, end=",")

def createTree(start_i, end_i): 
    # start_i, end_i: start and end indices for inorder data
    global preOrderIndex # next root index in preorder data 
    if start_i > end_i:
        return None
    node = Node(preOrderData[preOrderIndex]) #create root node
    preOrderIndex+=1
    # get the root node index position in inorder data
    inOrderIndex = inOrderData.index(node.value, start_i, end_i+1)
    # recursively create the left and right subtrees
    node.left = createTree(start_i, inOrderIndex-1)
    node.right = createTree(inOrderIndex+1, end_i)
    return node

def traverseDFSPreOrder(root):
    if root:
        visitNode(root)
        traverseDFSPreOrder(root.left)
        traverseDFSPreOrder(root.right)
```

### Word Search 
- See reference https://leetcode.com/problems/word-search/description/
- Given an m x n grid of characters board and a string word, return true if word exists in the grid.
- The word can be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. 
- The same letter cell may not be used more than once.
- Input: board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCCED"
    - Output: true
- Input: board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "SEE"
    - Output: true
- Input: board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCB"
    - Output: false

- Constraints:
    - m == board.length
    - n = board[i].length
    - 1 <= m, n <= 6
    - 1 <= word.length <= 15
    - board and word consists of only lowercase and uppercase English letters.

```python
from typing import List
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        visited=[]
        for r in range(len(board)):
            for c in range(len(board[0])):
                if dfs(r,c,board,word,visited):
                    return True
        return False
        
def dfs(r,c,board,word,visited):
    if len(word)==0:
        return True
    if r<0 or r>=len(board) or c<0 or c>=len(board[0]):
        return False
    if (r,c) in visited:
        return False
    if word[0]==board[r][c]:
        visited.append((r,c))
        if dfs(r+1,c,board,word[1:],visited):
            return True
        if dfs(r-1,c,board,word[1:],visited):
            return True
        if dfs(r,c+1,board,word[1:],visited):
            return True
        if dfs(r,c-1,board,word[1:],visited):
            return True
        visited.pop()
    return False

board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
# word = "ABCCED"
# word = "SEE"
word = "ABCB"
s = Solution()
s.exist(board,word)
```
