```python
# implements a BST(Binary Search Tree)
# Define a class Node which has insert() and search() instance methods
class Node:
    def __init__(self, key):
        self.key = key
        self.left=None
        self.right=None
    def insert(self, key):
        if key<self.key:
            if self.left!=None:
                return self.left.insert(key)
            else:
                self.left = Node(key)
        elif key>self.key:
            if self.right!=None:
                return self.right.insert(key)
            else:
                self.right = Node(key)
        else:
            assert False
    def search(self, key):
        if key == self.key:
            return self
        elif key<self.key:
            if self.left!=None:  
                return self.left.search(key)
            else:
                return None
        else:
            if self.right!=None:  
                return self.right.search(key)
            else:
                return None
    def printNode(self):
        if self.left:
            self.left.printNode()
        print(self.key,end=',')
        if self.right:
            self.right.printNode()
    def toList(self):
        result = [] 
        if self.left:
            result+=(self.left.toList()) # or use extend()
        result.append(self.key)
        if self.right:
            result.extend(self.right.toList())
        return result            
def main():
    data =[50,30,20,40,70,60,80,10]
    root = Node(data[0])
    for d in data[1:]:
        root.insert(d)
    root.printNode()
    print()
    result = root.toList()
    print(result)
    print(root.search(6))
    print(root.search(60).key)

main()
```

```python
# another implementation
# define Node class and general methods insert() and search()

class Node:
    def __init__(self, key):
        self.key = key
        self.left=None
        self.right=None
        
def insert(node, key):
    if node==None:
        return Node(key)
    elif key<node.key:
        node.left = insert(node.left,key)
    elif key>node.key:
        node.right = insert(node.right,key)
    else:
        assert False
    return node

def search(node,key):
    if node==None:
        return None
    elif key<node.key:
        return search(node.left,key)
    elif key>node.key:
        return search(node.right,key)
    else:
        return node

def printNode(node):
    if not node:
        return
    printNode(node.left)
    print(node.key,end=',')
    printNode(node.right)
        
def toList(node):
    result = []
    if not node:
        return result
    result+=(toList(node.left)) # or use extend()
    result.append(node.key)
    result.extend(toList(node.right))
    return result
def main():
    data = [50,30,20,40,70,60,80,10]
    root = None
    root = insert(root, data[0])
    for d in data[1:]:
        insert(root, d)
    printNode(root)    
    print()
    result = toList(root)
    print(result)
    print(search(root, 6))
    print(search(root, 60).key)

main()
```
