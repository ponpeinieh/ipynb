```python
# zerojudge m371. 2. 卡牌遊戲 2023年10月APCS
def scoreAny(r,c,value):
    global score
    # check up
    for i in range(r-1,-1,-1):
        if data[i][c]==value:
            score+=value
            data[r][c]=-1
            data[i][c]=-1
            return True
        elif data[i][c]==-1:
            continue
        else:
            break
    # check down
    for i in range(r+1,n):
        if data[i][c]==value:
            score+=value
            data[r][c]=-1
            data[i][c]=-1
            return True
        elif data[i][c]==-1:
            continue
        else:
            break
    # check left
    for i in range(c-1,-1,-1):
        if data[r][i]==value:
            score+=value
            data[r][c]=-1
            data[r][i]=-1
            return True
        elif data[r][i]==-1:
            continue
        else:
            break
    # check right
    for i in range(c+1,m):
        if data[r][i]==value:
            score+=value
            data[r][c]=-1
            data[r][i]=-1
            return True
        elif data[r][i]==-1:
            continue
        else:
            break
    traversed.append(value)
    return False
def calc():
    global traversed
    foundMore=False
    while True:
        for r in range(n):
            for c in range(m):
                if data[r][c]!=-1 and data[r][c] not in traversed and scoreAny(r,c,data[r][c]):
                    foundMore=True
        if not foundMore:
            break
        traversed=[]   
        foundMore=False
                
n,m = map(int, input().split())
data = [list(map(int, input().split())) for _ in range(n)]
# n,m=1,8
# data=[[0,2,3,3,0,2,5,5]]
# n,m=3,6
# data=[[0,2,3,8,0,2],[1,1,4,4,5,7],[5,6,3,8,6,7]]
score=0
traversed=[]
calc()
print(score)
# print(data)
```
