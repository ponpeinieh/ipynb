```python
# Binary Heaps (Max Heap)

# To heapify a subtree rooted with node i which is
# an index of arr[] and n is the size of heap
def heapify_subtree(arr, i):
	largest = i #Initialize largest as root
	l = 2 * i + 1 # left child
	r = 2 * i + 2 # right child

	#If left child is larger than root
	if l < len(arr) and arr[l] > arr[largest]:
		largest = l

	#If right child is larger than largest so far
	if r < len(arr) and arr[r] > arr[largest]:
		largest = r

	# If largest is not root
	if (largest != i):
		arr[i],arr[largest]=arr[largest],arr[i] #swap
		#Recursively heapify the affected sub-tree
		heapify_subtree(arr, largest)
#Function to delete the root from Heap
def deleteRoot(arr):
	# Replace root with last element
	arr[0] = arr[len(arr) - 1]
	# remove last element
	del arr[len(arr)-1]
	# heapify the root node
	heapify_subtree(arr, 0)

def heapify_parent(arr, i):
	parent = int((i-1)/2)
	# If current node is greater than its parent
	# Swap them and call heapify for the parent
	if parent >= 0:
		if arr[i] > arr[parent]:
			arr[i], arr[parent] = arr[parent], arr[i] #swap
			# Recursively heapify the parent node
			heapify_parent(arr, parent)
            
# Function to insert a new node to the Heap
def insertNode(arr, key):
	# Insert the element at end of Heap
	arr.append(key)
	# Heapify the new node following a Bottom-up approach
	heapify_parent(arr, len(arr)-1)
    
# A utility function to print array of size n
def printArray(arr):
	for e in arr:
		print(e,end=",") 
	print() 
 
arr = []
# 10, 3, 2, 5, 4, 11
insertNode(arr, 10)
printArray(arr)
insertNode(arr, 3)
printArray(arr)
insertNode(arr, 2)
printArray(arr)
insertNode(arr, 5)
printArray(arr)
insertNode(arr, 4)
printArray(arr)
insertNode(arr, 11)
printArray(arr)
#
deleteRoot(arr)
printArray(arr)
deleteRoot(arr)
printArray(arr)
```

```python
# use python module : heapq
# note that heapq implements the Min Heap, so in order to use it as a Max Heap we multiple the key by -1
def printArray(arr):
	for e in arr:
		print(-1*e,end=",") 
	print() 
import heapq
arr = []
# 10, 3, 2, 5, 4, 11
heapq.heappush(arr, -10)
printArray(arr)
heapq.heappush(arr, -3)
printArray(arr)
heapq.heappush(arr, -2)
printArray(arr)
heapq.heappush(arr, -5)
printArray(arr)
heapq.heappush(arr, -4)
printArray(arr)
heapq.heappush(arr, -11)
printArray(arr)
heapq.heappop(arr)
printArray(arr)
heapq.heappop(arr)
printArray(arr)
```
