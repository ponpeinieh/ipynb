```python
# 2024年1月APCS Q4
# use DP top-down
# also print the min cost path
def dfs(data, N):
    if N==1:
        return (0,(data,)) # 0 cost, also return the path with the min cost(a tuple)
    hashcode = hash(data)
    if use_lookup_table and hashcode in lookupTable: 
        return lookupTable[hashcode]   
    local_min_cost = 2000*(N-1)
    path_min_cost=None
    for i in range(N-1):
        arr = []
        for j in range(N):
            if j==i:
                arr.append(data[j]+data[j+1]) # sum of elements of indices j and j+1
                current_cost = abs(data[j]-data[j+1]) # the cost for the summation
                continue
            if j==i+1:
                continue
            arr.append(data[j])
        sub_cost, path = dfs(tuple(arr), N-1) 
        if sub_cost+current_cost<local_min_cost:
            path_min_cost = path 
            local_min_cost = sub_cost+current_cost
    path_min_cost = (data,) + path_min_cost
    if use_lookup_table:
        lookupTable[hashcode]=(local_min_cost, path_min_cost) 
    return (local_min_cost,path_min_cost)

# N = int(input())
# data = tuple(map(int,input().split()))
N = 10
# data = (-2, 1, 3, 2, 4)
import random
random.seed(3)
data = tuple(random.sample(range(1,21),k=N))
print(data)
lookupTable=dict()
use_lookup_table=True
print(dfs(data, N)) # data, size
```

```python
# 2024年1月APCS Q4
# DFS backtrack, no DP
# implementation 3: dfs() return mimimum cost for the current node (ie. sum of the top-down cost to the node and the minimum cost for the sub tree)
def dfs(data, N, cost):  # cost: the top-down total cost so far
    if N==1:
        return cost
    local_min_cost = 2000*(N-1)
    for i in range(N-1):
        arr = []
        for j in range(N):
            if j==i:
                arr.append(data[j]+data[j+1])
                current_cost = abs(data[j]-data[j+1])
                continue
            if j==i+1:
                continue
            arr.append(data[j])
        sub_cost = dfs(arr, N-1, cost+current_cost) 
        if sub_cost<local_min_cost:
            local_min_cost = sub_cost
    return local_min_cost

# N = int(input())
# data = list(map(int,input().split()))
N = 5
data = [-2, 1, 3, 2, 4]
arr = []
print(dfs(data, N, 0)) # data, size, cost
```

```python
# 2024年1月APCS Q4
# DFS iteration with a stack
def dfs_iterative():
    global min_cost
    stack = []
    for i in range(N-1): # 0 ~ N-2
        arr = []
        cost = 0
        for j in range(N):
            if j==i:
                arr.append(data[j]+data[j+1])
                cost = abs(data[j]-data[j+1])
                continue
            if j==i+1:
                continue
            arr.append(data[j])
        stack.append((arr,cost)) # store the arr data and cost so far
    #print(stack)
    while stack:
        element = stack.pop()
        parent_data = element[0]
        cost = element[1]
        M = len(parent_data)
        if M==1:
            if cost<min_cost:
                min_cost = cost
            continue
        for i in range(M-1): # 0 ~ M-2
            arr = []
            current_cost = 0
            for j in range(M):
                if j==i:
                    arr.append(parent_data[j]+parent_data[j+1])
                    current_cost = abs(parent_data[j]-parent_data[j+1])
                    continue
                if j==i+1:
                    continue
                arr.append(parent_data[j])
            if current_cost+cost>min_cost: #  add to the stack only if it's less than min_cost
                continue
            stack.append((arr,current_cost+cost))  # store the arr data and cost so far
# N = int(input())
# data = tuple(map(int,input().split()))
N = 15
# data = (-2, 1, 3, 2, 4)
import random
random.seed(3)
data = tuple(random.sample(range(1,21),k=N))
print(data)

min_cost = 2000*(N-1)
dfs_iterative()
print(min_cost)
```
