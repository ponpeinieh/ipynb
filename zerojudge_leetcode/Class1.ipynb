```python
# zerojudge no.17, a038, page 1
def reverse(x):
    result = 0
    while x:
        result=result*10+x%10
        x=x//10
    return result

x = int(input())
print(reverse(x))
```

```python
# zerojudge no.18, a040, page 1
def digits(i):
    c=0
    while i!=0:
        i=i//10
        c+=1
    return c
def amstrong(n,m):
    found = 0
    for i in range(n,m+1):
        a = 0
        c = digits(i)
        j = i
        while j!=0:
            a = a + (j%10)**c
            j=j//10
        if i == a:
            found = 1
            print(i, end=" ")
    if not found:
        print('none')
n,m = map(int, input().split())
amstrong(n,m)
```

```python
# zerojudge no.21, a053, page 2
'''
答對題數在 0~10 者，每題給6分。
題數在 11~20 者，從第11題開始，每題給2分。(前10題還是每題給6分)
題數在 21~40 者，從第21題開始，每題給1分。
題數在 40 以上者，一律100分。
'''
def score(n):
    if n<=10:
        return n*6
    elif n<=20:
        return 10*6+(n-10)*2
    elif n<=40:
        return 10*6+10*2+(n-20)
    else:
        return 100

print(score(int(input())))
```

```python
# zerojudge no.14, a022, page 1
def palindrome(mystr):
    for i in range(len(mystr)//2):
        if mystr[i] != mystr[len(mystr)-i-1]:
            print('no')
            return
    print('yes')

palindrome(input())
```

```python
# zerojudge no.24, a059, page 2
import math
def square_sum(n,m):
    sum=0
    for i in range(math.floor(math.sqrt(n)), math.floor(math.sqrt(m))+2):
        if n<=i**2<=m:
            sum+=(i**2)
    return sum

num = int(input())
for i in range(num):
    n = int(input())
    m = int(input())
    print(f'Case {i+1}: {square_sum(n,m)}')
```

```python
# zerojudge no.25, a065, page 2
def encode_password(my_str):
    passwd=[]
    for i in range(len(my_str)):
        if i+1<len(my_str):
            passwd.append(abs(ord(my_str[i])-ord(my_str[i+1])))
    for n in passwd:
        print(n,end='')
encode_password(input())
```

```python
# bubble sort : try to reduce the number of comparison
def bubble_sort(data):
    count=0
    for i in range(len(data)-1, 0, -1):
        for j in range(1, i+1):
            count+=1
            if data[j-1]>data[j]:
                data[j-1],data[j]= data[j],data[j-1] #swap
    print(count)

    
# Note that below optimized bubble sort uses 'last_swap' 
# to record the last swap performed, ie. it means beyond 
# last_swap, the elements have already sorted.
# It can not be done using for-in loop, since we are not 
# able to change the i to last_swap-1 in the middle of the loop
# So use while loop instead...
def bubble_sort2(data):
    count=0
    i = len(data)-1
    while i>0:
        last_swap=0
        for j in range(1, i+1):
            count+=1
            if data[j-1]>data[j]:
                data[j-1],data[j]= data[j],data[j-1] #swap
                last_swap = j
        i=last_swap-1
    print(count)

while True:
    try:
        n = int(input())
        data = list(map(int, input().split()))
        bubble_sort2(data)
        for n in data:
            print(n, end=' ')
        print()
    except EOFError:
        break
        
```

```python
# 埃拉托斯特尼篩法（sieve of Eratosthenes）,埃氏篩,質數篩可以用來找出一定範圍內所有質數。
# see reference: https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes
import math
def primes_within(k):
    is_prime = [True]*(k+1)
    is_prime[0]=is_prime[1]=False
    for i in range(2, math.floor(math.sqrt(k))+1):
        if is_prime[i]:
            for j in range(i*i, k+1, i):
                is_prime[j]=False
    return [i for i in range(k+1) if is_prime[i]]
K = 2147483647
primes = primes_within(math.floor(math.sqrt(K)))
def is_prime_sieve(n):
    is_prime=True
    for p in primes:
        if p*p > n:
            break
        elif n%p==0 and n!=p:
            is_prime=False
            break
    return is_prime


while True:
    try:
        n = int(input())
        if is_prime_sieve(n):
            print('質數')
        else:
            print('非質數')
    except EOFError:
        break
```
