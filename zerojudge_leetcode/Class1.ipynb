```python
# zerojudge no.18, a040, page 1
def digits(i):
    c=0
    while i!=0:
        i=i//10
        c+=1
    return c
def amstrong(n,m):
    found = 0
    for i in range(n,m+1):
        a = 0
        c = digits(i)
        j = i
        while j!=0:
            a = a + (j%10)**c
            j=j//10
        if i == a:
            found = 1
            print(i, end=" ")
    if not found:
        print('none')
n,m = map(int, input().split())
amstrong(n,m)
```

```python
# zerojudge no.21, a053, page 2
'''
答對題數在 0~10 者，每題給6分。
題數在 11~20 者，從第11題開始，每題給2分。(前10題還是每題給6分)
題數在 21~40 者，從第21題開始，每題給1分。
題數在 40 以上者，一律100分。
'''
def score(n):
    if n<=10:
        return n*6
    elif n<=20:
        return 10*6+(n-10)*2
    elif n<=40:
        return 10*6+10*2+(n-20)
    else:
        return 100

print(score(int(input())))
```

```python
# bubble sort : try to reduce the number of comparison
def bubble_sort(data):
    count=0
    for i in range(len(data)-1, 0, -1):
        for j in range(1, i+1):
            count+=1
            if data[j-1]>data[j]:
                data[j-1],data[j]= data[j],data[j-1] #swap
    print(count)

    
# Note that below optimized bubble sort uses 'last_swap' 
# to record the last swap performed, ie. it means beyond 
# last_swap, the elements have already sorted.
# It can not be done using for-in loop, since we are not 
# able to change the i to last_swap-1 in the middle of the loop
# So use while loop instead...
def bubble_sort2(data):
    count=0
    i = len(data)-1
    while i>0:
        last_swap=0
        for j in range(1, i+1):
            count+=1
            if data[j-1]>data[j]:
                data[j-1],data[j]= data[j],data[j-1] #swap
                last_swap = j
        i=last_swap-1
    print(count)

while True:
    try:
        n = int(input())
        data = list(map(int, input().split()))
        bubble_sort2(data)
        for n in data:
            print(n, end=' ')
        print()
    except EOFError:
        break
        
```

```python
# zerojudge 原創/不分類題庫 no.1, a007, page 1
# Time out!!!
#
# 埃拉托斯特尼篩法（sieve of Eratosthenes）,埃氏篩,質數篩可以用來找出一定範圍內所有質數。
# see reference: https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes
import math
def primes_within(k):
    is_prime = [True]*(k+1)
    is_prime[0]=is_prime[1]=False
    for i in range(2, math.floor(math.sqrt(k))+1):
        if is_prime[i]:
            for j in range(i*i, k+1, i):
                is_prime[j]=False
    return [i for i in range(k+1) if is_prime[i]]
K = 2147483647
primes = primes_within(math.floor(math.sqrt(K)))
def is_prime_sieve(n):
    is_prime=True
    for p in primes:
        if p*p > n:
            break
        elif n%p==0 and n!=p:
            is_prime=False
            break
    return is_prime


while True:
    try:
        n = int(input())
        if is_prime_sieve(n):
            print('質數')
        else:
            print('非質數')
    except EOFError:
        break
```

```python
# zerojudge 原創/不分類題庫 no.1, a007, page 1
# Time out!!!
# Miller - Rabin primality test
import random
def power(x,y,p):
    result=1
    while y>0:
        if y%2:
            result *= (x % p)
        y = y>>1
        x = (x * x) % p
    return result % p

def miller_test(d,n):
    a = 2 + random.randint(1, n-4)
    x = power(a,d,n)
    if x==1 or x==n-1:
        return True
    while d<n-1:
        x = (x * x) % n
        d*=2
        if x==1:
            return False
        if x==n-1:
            return True
    return False

def is_prime(n, k):
    if n<=1 or n==4:
        return False
    if n<=3:
        return True
    d = n-1
    while d%2==0:
        d//=2
    for i in range(k):
        if miller_test(d,n)==False:
            return False
    return True

while True:
    try:
        n = int(input())
        if is_prime(n,4):
            print('質數')
        else:
            print('非質數')
    except EOFError:
        break
            
```

```python
# zerojudge no.44, a417
def calc_array(N,M):
    d = [[0]*N for _ in range(N)]
    s1 = [] # store [size of each side, sign(1 for increase, -1 decrease)]
    s2 = 1
    for i in range(N,0,-1): #append N,N-1,N-1,N-2,N-2,...,1,1
        s1.append([i])
        if i!=N:
            s1.append([i])
    for i in range(0,len(s1),2): # append sign(1 for increase, -1 decrease), ie. 1,1,-1,-1,1,1,...
        s1[i].append(s2)
        if i+1>=len(s1):
            break
        s1[i+1].append(s2)
        s2*=-1
        
    x = 1 # 1 to N*N
    if M==1: #clockwize, first along row
        dn=1 #'c right'
        t=[0,-1]
    elif M==2: #counter clockwize , first along column 
        dn=-1 #'r down'
        t=[-1,0]
    for m in s1:
        for n in range(m[0]): 
            if dn==1:
                t[1]+=m[1]
            elif dn==-1:
                t[0]+=m[1]
            d[t[0]][t[1]]=x
            x+=1
        dn*=-1
    return d

def print_element(i):
    print("{0: >5}".format(i),end="")
    
T = int(input())
for i in range(T):
    N,M = map(int,input().split())
    d = calc_array(N,M)
    for r in d:
        for c in r:
            print_element(c)
        print()
```
