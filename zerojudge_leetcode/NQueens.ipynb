```python
# N Queens problem
def printBoards():
    for r in boards:
        for c in r:
            if c:
                print('Q', end=" ")
            else:
                print('.',end=" ")
        print()
def validPosition(r,c):
    # check same row?
    for i in range(c):
        if boards[r][i]:
            return False
    # check diagonals
    for i,j in zip(range(r-1,-1,-1),range(c-1,-1,-1)):
        if boards[i][j]:
            return False
    for i,j in zip(range(r+1,N),range(c-1,-1,-1)):
        if boards[i][j]:
            return False        
    return True
count=0
def NQueens(L):
    global count
    count+=1
    if L==N:
        return True
    for r in range(N):
        if validPosition(r,L):
            boards[r][L]=1
            if NQueens(L+1):
                return True
            boards[r][L]=0
    return False

N = 10
boards = [[0]*N for _ in range(N)]
L=0
NQueens(L)
printBoards()
print(count)
```
